```python
fig = plt.figure(figsize=(30, 20))
ax = fig.add_subplot(111)

ph.plot(ax=ax, color='white', edgecolor='dimgray')
healthsites.plot(ax=ax, column="amenity", alpha=0.35, legend=True)
ax.set_title('Health-related Amenities Near Me', fontsize=15)
ax.axis('off');
```

```python
fig = plt.figure(figsize=(30, 20))
ax = fig.add_subplot(111)

ph.plot(ax=ax, color='white', edgecolor='dimgray')
healthsites.plot(ax=ax, column="amenity", cmap="Paired", alpha=0.35, legend=True)
ax.set_title('Health-related Amenities in the Philippines', fontsize=15)
ax.axis('off');
```

```python
mm = ph[ph.NAME_1=="Metropolitan Manila"]
bl = ph[ph.NAME_1=="Bulacan"]
mmbl = mm.append(bl)
mmbl.plot()
```

```python
amenities_within = gpd.sjoin(healthsites, mmbl, how='inner', op='within')
amenities_within.head()
```

```python
fig = plt.figure(figsize=(15, 20))
ax = fig.add_subplot(111)

mmbl.plot(ax=ax, color='white', edgecolor='dimgray')
amenities_within.plot(ax=ax, column="amenity", cmap="Paired", alpha=0.35, legend=True)
ax.set_title('Health-related Amenities Within Metro Manila and Bulacan', fontsize=15)
ax.axis('off');
```

```python
fig = plt.figure(figsize=(30, 20))
ax = fig.add_subplot(111)

mmbl.plot(ax=ax, color='white', edgecolor='dimgray')
amenities_within.loc[amenities_within['amenity'] == 'dentist'].plot(ax=ax, column="amenity", cmap="Paired", legend=True)
ax.set_title('Dentist Amenities Within Metro Manila and Bulacan', fontsize=15)
ax.axis('off');
```
